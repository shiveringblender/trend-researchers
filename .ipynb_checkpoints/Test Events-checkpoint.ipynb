```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from alpha_vantage.timeseries import TimeSeries
api_key = 'YZFIGXM1LVE5BOOC'
```

```python
stock_symbols = ['AAPL', 'GOOGL', 'MSFT', 'AMZN']
event_date = pd.to_datetime('2020-03-15')
window_size = 52
end_date = pd.to_datetime(event_date)
start_date = end_date - pd.DateOffset(years=1)
one_year_after_date = end_date + pd.DateOffset(years=1)
```

```python
ts = TimeSeries(key=api_key, output_format='pandas')
stock_data_dict = {}
```

```python
for symbol in stock_symbols:
    data, meta_data = ts.get_weekly_adjusted(symbol=symbol)
    
    # Ensure that the date index is sorted in ascending order
    data = data.sort_index()
    
    # Filter data for the year before and after the event
    selected_data = data[(data.index >= start_date) & (data.index <= one_year_after_date)]
    
    # Resample data to weekly frequency (in case it's not already)
    selected_data = selected_data.resample('W').last()
    
    stock_data_dict[symbol] = selected_data
```

```python
avg_price_before_event_dict = {}
avg_price_after_event_dict = {}
std_dev_before_event_dict = {}
std_dev_after_event_dict = {}

for symbol, data in stock_data_dict.items():
    avg_price_before_event = data[data.index < event_date]['5. adjusted close'].mean()
    std_dev_before_event = data[data.index < event_date]['5. adjusted close'].std()
    
    avg_price_after_event = data[data.index >= event_date]['5. adjusted close'].mean()
    std_dev_after_event = data[data.index >= event_date]['5. adjusted close'].std()
    
    avg_price_before_event_dict[symbol] = avg_price_before_event
    std_dev_before_event_dict[symbol] = std_dev_before_event
    avg_price_after_event_dict[symbol] = avg_price_after_event
    std_dev_after_event_dict[symbol] = std_dev_after_event
```

```python
for symbol, data in stock_data_dict.items():
    print(f"Stock Symbol: {symbol}\n{data}\n")
```

```python
symbols = list(stock_symbols)
avg_prices_before_event = [avg_price_before_event_dict[s] for s in stock_symbols]
avg_prices_after_event = [avg_price_after_event_dict[s] for s in stock_symbols]
```

```python
std_dev_before_event = [std_dev_before_event_dict[s] for s in stock_symbols]
std_dev_after_event = [std_dev_after_event_dict[s] for s in stock_symbols]
```

```python
bar_width = 0.35
index = range(len(symbols))
bar_positions_before = [i - bar_width/2 for i in index]
bar_positions_after = [i + bar_width/2 for i in index]
```

```python
plt.figure(figsize=(12, 6))
plt.bar(bar_positions_before, avg_prices_before_event, width=bar_width, label='One Year Before Event', alpha=0.7, yerr=std_dev_before_event, capsize=5)
plt.bar(bar_positions_after, avg_prices_after_event, width=bar_width, label='One Year After Event', alpha=0.7, yerr=std_dev_after_event, capsize=5)
plt.xlabel('Stock Symbols')
plt.ylabel('Average Adjusted Closing Price($)')
plt.title('Stock Price Comparison Before and After COVID-19 Lockdowns')
plt.legend()
plt.xticks(index, symbols)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for symbol, data in stock_data_dict.items():
    plt.plot(data.index, data['5. adjusted close'], label=symbol)

plt.axvline(event_date, color='red', linestyle='--', label='Event Date')
plt.xlabel('Date')
plt.ylabel('Adjusted Closing Price($)')
plt.title('Historical Stock Prices (One Year Before and After Event)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for symbol, data in stock_data_dict.items():
    std_dev = data['5. adjusted close'].rolling(window=window_size, min_periods=1).std()
    plt.plot(data.index, std_dev, label=f'{symbol} Std. Dev.')
    
plt.axvline(event_date, color='red', linestyle='--', label='Event Date')
plt.xlabel('Date')
plt.ylabel('Standard Deviation of Adjusted Close Price')
plt.title('Price Volatility Before and After Event')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for symbol, data in stock_data_dict.items():
    returns = (data['5. adjusted close'] / data['5. adjusted close'].iloc[0] - 1) * 100
    plt.plot(data.index, returns, label=f'{symbol} Cumulative Returns')

plt.axvline(event_date, color='red', linestyle='--', label='Event Date')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns (%)')
plt.title('Cumulative Returns Before and After Event')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
